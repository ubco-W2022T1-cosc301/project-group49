```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.pylab as plt

df=pd.DataFrame({'Country': ["Uruguay", "Italy", "France", "France", "Uruguay"], 'Winner': ["Uruguay", "Italy", "Italy", "France", "Italy"]})

df
```

```python
def f(row):
    if row['Country'] in row['Winner']:
        val = False
    else:
        val = True
    return val

df["AwayTeam"]=df.apply(f, axis=1)
df
```

```python
df.drop(df[df['AwayTeam'] == False].index, inplace = True)
df = df.drop("AwayTeam", axis=1)
df
```
