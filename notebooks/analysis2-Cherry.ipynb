```python
df['Year'] = df['Year'].astype('int64')
df['GoalsScored'] = df['GoalsScored'].astype('int64')
df['QualifiedTeams'] = df['QualifiedTeams'].astype('int64')
df['MatchesPlayed'] = df['MatchesPlayed'].astype('int64')

df['Country'] = df['Country'].astype('str')
df['Winner'] = df['Winner'].astype('str')
df['Runners-Up'] = df['Runners-Up'].astype('str')
df['Third'] = df['Third'].astype('str')
df['Fourth'] = df['Fourth'].astype('str')

df['Attendance'] = df['Attendance'].astype('str')

df.info()
```

```python
DeleteList=['Year','Runners-Up','Third','Fourth','GoalsScored','QualifiedTeams','MatchesPlayed','Attendance']
df = df.drop(DeleteList, axis=1)
df.head()
```

```python
def f(row):
    if row['Country'] in row['Winner']:
        val = False
    else:
        val = True
    return val
```

```python
df['AwayTeam'] = df.apply(f, axis=1)
df
```

```python
df.drop(df[df['AwayTeam'] == False].index, inplace = True)
df = df.drop("AwayTeam", axis=1)
df
```

```python
df = df.drop("AwayTeam", axis=1)
df
```
